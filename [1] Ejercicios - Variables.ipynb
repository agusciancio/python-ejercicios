# Ejercicios - Variables


















## Ejercicio 1: Hola, mundo
Escribe un programa que imprima en pantalla el mensaje:  
`Hola, mundo`.

---

## Ejercicio 2: Declaración de variables
Declara las siguientes variables y asigna los valores correspondientes:
- Una variable `nombre` con tu nombre.
- Una variable `edad` con tu edad.
- Una variable `altura` con tu altura en metros.

Imprime cada una de las variables.

---

## Ejercicio 3: Operaciones con números
Declara dos variables, `a` y `b`, con valores 10 y 5, respectivamente.  
Calcula e imprime los resultados de las siguientes operaciones:
1. Suma
2. Resta
3. Multiplicación
4. División

---

## Ejercicio 4: Concatenación de cadenas
Crea dos variables, `saludo` y `nombre`, con los valores `"Hola"` y tu nombre.  
Usa estas variables para imprimir un mensaje de bienvenida como:  
`"Hola Tu_Nombre"`

---

## Ejercicio 5: Tipos de datos
Declara una variable de cada uno de estos tipos:
- Entero (`int`)
- Punto flotante (`float`)
- Cadena (`str`)
- Booleano (`bool`)

Imprime el tipo de dato de cada variable usando `type()`.

---

## Ejercicio 6: Conversión de tipos
Declara una variable `edad` como un número entero.  
Convierte esa variable en una cadena y luego imprime el mensaje:  
`"Tienes X años"`  
(Sustituye `X` por la edad).

---

## Ejercicio 7: Intercambio de valores
Declara dos variables `x` y `y` con valores 5 y 10, respectivamente.  
Intercambia los valores de las variables y luego imprime sus nuevos valores.

---

## Ejercicio 8: Operaciones avanzadas
Declara una variable `radio` con el valor 7.  
Calcula e imprime el área de un círculo usando la fórmula:  
`Área = π * radio^2`  
(Usa 3.14 para el valor de π).

---

## Ejercicio 9: Entrada del usuario
Escribe un programa que:
1. Solicite al usuario su nombre.
2. Solicite su edad.
3. Imprima un mensaje como:  
   `"Hola, [nombre]. Tienes [edad] años."`

---

## Ejercicio 10: Cálculo de promedio
Declara tres variables `nota1`, `nota2` y `nota3` con valores de tres calificaciones.  
Calcula e imprime el promedio de las tres notas.

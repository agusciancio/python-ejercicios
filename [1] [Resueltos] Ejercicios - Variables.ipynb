## Ejercicio 6: Conversión de tipos
Declara una variable `edad` como un número entero.

`"Tienes X años"`  
(Sustituye `X` por la edad).

## Ejercicio 7: Intercambio de valores

## Ejercicio 8: Operaciones avanzadas
Declara una variable `radio` con el valor 7.  
Calcula e imprime el área de un círculo usando la fórmula:  
`Área = π * radio^2`

## Ejercicio 10: Cálculo de promedio
Declara tres variables `nota1`, `nota2` y `nota3` con valores de tres calificaciones.  
Calcula e imprime el promedio de las tres notas.

```python
nota1 = 10
nota2 = 8
nota3 = 7
promedio = (nota1 + nota2 + nota3) / 3
print(f'El promedio es {round(promedio, 2)}')
```
